# Working with API's

### What is an API??

- Application Programming Interface (API)
- These are data pipelines, which essentially make two softwares talk 

Consider, a train ticket booking scenario with IRCTC, makemytrip, ixigo etc. If you book a ticket through IRCTC, the number of avaiable seats gets automatically updated even in makemytrip. How does makemytrip know that these number of seats were booked??

The Answer is APIs,

IRCTC has created APIs which these private companies can use to return a JSON file. This will then update the database and we get the results we see on the screens.

### TMDB website

We are using this website for our exercise. 

- Search "tmdb api" and you'll find it
- What this essentially is, is a URL
- We use this URL and your API key to get the data
- To get the API key make an account in the website and fill in some creds
- make sure to add "?api_key=<'yourkey'>&" before language in the URL to get the JSON file

```python
import pandas as pd
import requests
```

```python
response = requests.get("https://api.themoviedb.org/3/tv/top_rated?api_key=7c871f5c30a3fd822952a0299a313e92&language=en-US&page=1?")
# response [200] means everything went smoothly
```

```python
response.json()['results']
```

```python
# to read this into a dataframe.....
pd.DataFrame(response.json()['results'])[['id', 'name', 'overview', 'first_air_date', 'popularity', 'vote_average', 'vote_count']].head()
# to read all the pages of json we have to apply loops
# make sure to lookup the number of pages from the JSON viewer online!!
```

```python
# create an empty dataset
df = pd.DataFrame()
```

```python
for i in range(1, 145):
    # call the api with fstring setup for pagenumber
    response = requests.get(f"https://api.themoviedb.org/3/tv/top_rated?api_key=7c871f5c30a3fd822952a0299a313e92&language=en-US&page={i}?")
    # extract rows from json and store in a temporary dataframe
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'name', 'overview', 'first_air_date', 'popularity', 'vote_average', 'vote_count']].head()
    # since the index of each page repeats, we keep ignore_index=True
    df = df.append(temp_df, ignore_index=True)
```

```python
print(df.shape)
# now you can export this to a CSV file
df.to_csv("movies.csv")
```
