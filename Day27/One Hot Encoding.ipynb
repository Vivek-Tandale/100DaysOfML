## One Hot Encoding

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
df
```

```python
# note that all brnads don't occur in the same frequency
df['brand'].value_counts()
```

```python
# fuel types are also not evenly distributed
df['fuel'].value_counts()
```

```python
# similarly for ownership
df['owner'].value_counts()
```

### 1. OneHotEncoding

```python
# basically we call this function to replace the categorical columns with the encoded ones
# we are ignoring the 'brands' for now
pd.get_dummies(df, columns=['fuel', 'owner'])
```

### 2. K-1 OneHotEncoding

```python
# removing the first coolumns to avoid multicolinearity
pd.get_dummies(df, columns=['fuel', 'owner'], drop_first=True)
# note that this has 2 columns less
```

The problem with this is Pandas doesn't remember which column was the first. It can remove a completly differnet column another time if runned.

Hence, we can use scikit learn instead which remebers the columns.

### 4. OneHotEncoding with Top Categories

```python
counts = df['brand'].value_counts()
# make a df which has counts of all the brand occurences
```

```python
df['brand'].nunique()
# this just says the number of unique brands
```

```python
# create a threshold
threshold = 100
```

```python
repl = counts[counts <= threshold].index
# get the index of all the brands with count greater than threshold
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon'))
# get_dummies but this time replace the non-frequent brands with just one column 'uncommon'
```
